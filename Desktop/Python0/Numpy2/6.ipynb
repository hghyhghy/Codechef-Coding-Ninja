```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,3*np.pi,0.1)

sinx = np.sin(x)
cosy = np.cos(x)

# plotting the points using matplotlib 

plt.plot(x,sinx)
plt.plot(x,cosy)

plt.xlabel("X plot")
plt.ylabel("Y plot")

plt.title("Sine and Cosine")
plt.legend(["Sine","cosine"])

plt.show()
```
