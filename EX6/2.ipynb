```python
import numpy as np
import matplotlib.pyplot as plt 

ypoint = np.array([3,8,1,10])

plt.plot(ypoint,'o:r')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt 

# use of marker size 

ypoint  = np.array([3,8,1,10])

plt.plot(ypoint, marker="o", ms=10)

plt.show()
```
